## What is COVID-19? 
COVID-19 is a respiratory disease caused by the Severe Acute Respiratory Coronavirus 2 (SARS-CoV-2). The virus is mainly transmitted via respiratory droplets (via talking, coughing, sneezing) or via contaminated objects (fomites). Symptoms include: 

- fever
- dry cough
- difficulty breathing
- fatigue
- muscle pain

The virus first originated from Wuhan, China - before it quickly spread all over the world. The World Health Organization (WHO) declared the outbreak as a 'pandemic' in March 2020. 

## Research Question

*My research question is 'How does COVID-19 mortality differ between various ethnic groups in the US?'*

I focused on COVID-19 data in all 50 US states on one day (October 18th 2020) to simplify the process. 

## Hypothesis 
I hypothesize that BIPOC (Black, Indigenous and People of Color) mortality will be disproportionately affected by COVID-19 due to the following reasons: 

- lower socioeconomic status
- less access to safe and reliable housing 
- poor access to healthcare
- increased occupational hazards
